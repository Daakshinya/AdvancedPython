```python
import seaborn as sns
import matplotlib.pyplot as plt
#load
tips=sns.load_dataset("tips")
#create a scatter plot
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title("scatter plot of total bill vs.Tip")
plt.xlabel("total bill ($)")
plt.ylabel("Tips($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#load
tips=sns.load_dataset("iris")
print(tips)
#create a scatter plot
sns.scatterplot(x="sepal_length",y="sepal_width",data=tips)
plt.title("scatter plot of sepal_length vs sepal_width")
plt.xlabel("sepal_width ($)")
plt.ylabel("sepal_length($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#load
tips=sns.load_dataset("iris")
print(tips)
#create a scatter plot
sns.scatterplot(x="sepal_length",y="sepal_width",data=tips)
#violinplot
sns.violinplot(x="sepal_length",y="sepal_width",data=tips)
plt.title("scatter plot of sepal_length vs sepal_width")
plt.xlabel("sepal_width ($)")
plt.ylabel("sepal_length($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset("iris")
print(iris)
correlation_matrix=iris.corr()
sns.heatmap(correlation_matrix,annot=True, cmap="coolwarm")
plt.title("correlation heatmamp of iris dataset")
plt.show()
```
