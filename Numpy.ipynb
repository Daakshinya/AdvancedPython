```python

import numpy as np
#creating array
arr=np.array([1,2,3,4,5])
print(arr)
a=np.zeros((3,3), dtype=int)
print(a)
b=np.ones((5,2), dtype=int)
print(b)

arange=np.arange(10)
print(arange)

#array manipulation
reshape=arr.reshape(5,1)
print(reshape)
#slicing
sliced_arr=arr[2:4]
print(sliced_arr)
#element wise operation
p=np.array([[1,2,3,4],[5,6,7,8]])
q=np.array([5,6,7,8])
result=p+q
print(result)

#stack format
a=np.array([1,2,3])
b=np.array([7,8,9])
c=np.stack((a+b))
print(c)

#splitting
x1=np.array([11,12,13,14,15,16])
result=np.split(x1,2)
print(result)
a=np.array([[1,2,3],[4,5,6]])
b=a.T
print(b)
```

```python
import numpy as np
#broad cast
a1=np.array([3,4,5])
a2=np.array([7,8,9])
result=a1+a2
print(result)
result1=1+a1
print(result1)
#transpose function
b=a1.T
print(b)
```

```python
import numpy as np
#linear algebra
a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])
c=np.dot(a,b)
print(c)
d=np.linalg.eig(c)
print(d)

a1=np.array([[3,4],[5,6]])
a2=np.array([[6,2],[1,3]])
c1=np.dot(a,b)
d1=c+c1
print(d1)

x=np.array([[1,2,3],[4,5,6]])
c=np.sum(x,axis=0)#row
d=np.sum(x,axis=1)#column
print(c)
print(d)

#Statistical operations
a=np.array([1,2,3,4,5])
mean=np.mean(a)
median=np.median(a)
variance=np.var(a)
standard=np.std(a)
print(mean)
print(median)
print(variance)
print(standard)
```
